```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import numpy as np
from sklearn.datasets import load_boston
boston_dataset = load_boston()
```

```python
boston_housing = pd.DataFrame(boston_dataset.data, columns=boston_dataset.feature_names)
boston_housing.columns
```

```python
boston_housing['MEDV'] = boston_dataset.target
```

```python
#X= boston_housing[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX','PTRATIO', 'B', 'LSTAT']]
X = pd.DataFrame(np.c_[boston_housing['CRIM'], boston_housing['ZN'],boston_housing['INDUS'], boston_housing['CHAS'],boston_housing['NOX'], boston_housing['RM'],boston_housing['AGE'], boston_housing['DIS'],boston_housing['RAD'], boston_housing['TAX'],boston_housing['PTRATIO'], boston_housing['B'],boston_housing['LSTAT']], columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX','PTRATIO', 'B', 'LSTAT'])

y=boston_housing[ 'MEDV']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=5)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.metrics import mean_squared_error

lin_model = LinearRegression()
lin_model.fit(X_train, y_train)
```

```python
y_pred=lin_model.predict(X_test)
```

```python
y_pred.round(decimals=2)
```

```python
lin_model.score(X_test,y_test)
```
