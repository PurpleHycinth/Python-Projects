```python
import numpy as np
import pandas as pd


df = pd.DataFrame(np.c_[np.arange(6), np.random.randint(1, 3, size=6)])  # make a dataset
def createrows(df,n_copies):
  subset = df[df.iloc[:, -1] > 1]   # getting the last column as a dataframe
  # making new rows
  extra = pd.concat([subset] * n_copies).add(np.random.uniform(0, 0.05, len(subset) * n_copies), axis='rows')
  # making a new data set
  result = df.append(extra, ignore_index=True)
  return result
```
