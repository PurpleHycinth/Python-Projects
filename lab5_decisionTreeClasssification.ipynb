```python
def probability(column):
  p= q =0
  for i in range(len(column)):
    if column[i] == 0: # 0 is "No"
      q +=1

    else :
      p +=1
  return (p/len(column)),(q/len(column))
```

```python
def gini_impurity(column):
  var = probability(column)
  giniImpurity = 1 - (var[0]) ** 2 - (var[1]) ** 2
  return giniImpurity
```

### Part B

```python

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import cross_val_score

le = LabelEncoder()


# split the dataset into training and testing sets
train['Sex'] = le.fit_transform(train['Sex'])
train['Embarked'] = le.fit_transform(train['Embarked'])
train = train.dropna(subset=['Age'])

X_train = train.drop(['PassengerId','Survived',"Name",'Ticket','Cabin'], axis=1)
y_train = train['Survived']
```

```python
Le = LabelEncoder()

test['Sex'] = Le.fit_transform(test['Sex'])
test['Embarked'] = Le.fit_transform(test['Embarked'])
test = test.dropna(subset=['Fare',"Age"])

X_test = test.drop(['PassengerId',"Name",'Ticket','Cabin',"Survived"], axis=1)
y_test = test['Survived']

```

```python
clf_G = DecisionTreeClassifier(criterion='gini',max_depth = 3,min_samples_split = 5)

# fit the classifier to the training data
clf_G.fit(X_train, y_train)
y_pred = clf_G.predict(X_test)
accuracy1 = accuracy_score(y_test, y_pred)
print(accuracy1)
```

```python
clf_G = DecisionTreeClassifier(criterion='gini',max_depth = 4,min_samples_split = 5)

# fit the classifier to the training data
clf_G.fit(X_train, y_train)
y_pred = clf_G.predict(X_test)
accuracy1 = accuracy_score(y_test, y_pred)
print(accuracy1)
```

```python
clf_L = DecisionTreeClassifier(criterion='log_loss',max_depth = 3,min_samples_split = 5)

# fit the classifier to the training data
clf_L.fit(X_train, y_train)
y_pred = clf_L.predict(X_test)
accuracy2 = accuracy_score(y_test, y_pred)
print(accuracy2)
```
